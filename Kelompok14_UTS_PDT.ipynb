Menghitung statistik data (Mean,median,mode) dari data yang ada

```python
import pandas as pd
import numpy as np
import sklearn
from scipy import stats
```

```python
df= pd.read_csv("https://raw.githubusercontent.com/2217051106/UTS-PDT/main/telecom_cus.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
#Nilai Rata-rata perkolom
mean = df.mean()
print("\nMean:")
print(mean)
```

```python
#Nilai rata-rata keseluruhan data
mean_all = df.mean().mean()
print("\nMean dari keseluruhan data:", mean_all)
```

```python
# Nilai median perkolom
median = df.median()
print("\nMedian:")
print(median)
```

```python
# Nilai median dari keseluruhan data
median_all = df.median().median()
print("\nMedian dari keseluruhan data:", median_all)
```

```python
# Nilai mode perkolom
mode = df.mode().iloc[0]
print("\nMode:")
print(mode)
```

```python
import matplotlib.pyplot as plt
from PIL import Image

# Misalkan file gambar yang diunggah bernama 'gambar.jpg'
img_path = 'foto kita.jpeg'

# Baca gambar
img = Image.open(img_path)

# Tampilkan gambar
plt.imshow(img)
plt.axis('off')  # Untuk menyembunyikan sumbu
plt.show()
```

```python
import cv2

# Load pre-trained foto classifier
foto_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_smile.xml')

# Ubah gambar menjadi array numpy
img_np = np.array(img)

# Konversi gambar ke skala abu-abu
gray = cv2.cvtColor(img_np, cv2.COLOR_BGR2GRAY)

# Deteksi objek dalam gambar
fotos = foto_classifier.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=2, minSize=(400, 400))

# Gambar kotak persegi panjang di sekitar objek yang terdeteksi
for (x, y, w, h) in fotos:
    cv2.rectangle(img_np, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Tampilkan gambar dengan objek yang terdeteksi
plt.imshow(img_np)
plt.axis('off')
plt.show()

# Print jumlah objek yang terdeteksi
print("Jumlah objek yang terdeteksi:", len(fotos))
```
